```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os, tqdm
import seaborn as sns
import scipy.io
import torch
from torch.utils.data import Dataset,DataLoader
import torch.nn as nn
import shutil
import matplotlib.pyplot as plt
```

```python

def loadmat(filepath):
    return scipy.io.loadmat(filepath, simplify_cells=True)
```

```python
import scipy.io
import os

folder_path = 'NASAdataset'
mat_files = [f for f in os.listdir(folder_path) if f.endswith('.mat')]

battery_data = {}  # Dictionary to hold datasets

for file in mat_files:
    full_path = os.path.join(folder_path, file)
    mat_data = scipy.io.loadmat(full_path)
    
    # Extract the dataset name, e.g., 'B0005' from the keys
    for key in mat_data:
        if key.startswith('B'):
            battery_data[key] = mat_data[key]
            print(f"Loaded {key} into battery_data['{key}']")
```

```python
b0005_data = battery_data['B0005'][0, 0]
print(len(b0005_data))
# cycle_data = b0005_data['cycle'][0][0]['data']
cycle_data = b0005_data['cycle']
```

```python
# Flatten the cycle_data to ensure it's in a 1D array for easy searching
cycle_types = cycle_data['type'].flatten()

# Count how many 'charge' cycles there are
charge_count = sum(1 for cycle in cycle_types if cycle == 'charge')

# Print the count
print("Number of 'charge' cycles:", charge_count)
```

```python
cycle_types = cycle_data['type'].flatten()

# Find indices of 'charge' cycles
charge_indices = [i for i, t in enumerate(cycle_types) if t == 'charge']

# Initialize lists to store extracted data for charge cycles
cycle_numbers = []
voltage_measured = []
current_measured = []
temperature_measured = []
current_charge = []
voltage_charge = []
time_data = []

# Loop through the charge cycles and extract the data
for idx in charge_indices:
    cycle =cycle_data[0, idx]['data']
    print(cycle.dtype.names)
    # # Extract relevant fields from the cycle
    voltage_measured.append(cycle['Voltage_measured'])
    current_measured.append(cycle['Current_measured'])
    temperature_measured.append(cycle['Temperature_measured'])
    current_charge.append(cycle['Current_charge'])
    voltage_charge.append(cycle['Voltage_charge'])
    time_data.append(cycle['Time'])
print("Number of 'charge' cycles:", len(charge_indices))
```

```python
# Assuming you already have cycle_data and the relevant fields extracted (like 'Voltage_measured', etc.)

# Initialize lists to store the extracted data
cycle_numbers = []
voltage_measured = []
current_measured = []
temperature_measured = []
current_charge = []
voltage_charge = []
time_data = []

# Loop through each cycle in cycle_data
for idx in range(len(cycle_data['type'])):
    cycle_type = cycle_data['type'][idx][0]  # Get the type (charge/discharge)
    
    if cycle_type == 'charge':  # Filter out only charge cycles
        # Extract relevant fields for each cycle
        cycle_numbers.append(idx + 1)  # Cycle number (1-based index)
        cycle =cycle_data[0, idx]['data']
        voltage_measured.append(cycle['Voltage_measured'][0][0][0].flatten())  # Voltage
        current_measured.append(cycle['Current_measured'][0][0])  # Current
        temperature_measured.append(cycle['Temperature_measured'][0][0])  # Temperature
        current_charge.append(cycle['Current_charge'][0][0])  # Current charge
        voltage_charge.append(cycle['Voltage_charge'][0][0])  # Voltage charge
        time_data.append(cycle['Time'][0][0])  # Time

# Convert the extracted data into a DataFrame
df = pd.DataFrame({
    'Cycle Number': cycle_numbers,
    'Voltage_measured': voltage_measured,
    'Current_measured': current_measured,
    'Temperature_measured': temperature_measured,
    'Current_charge': current_charge,
    'Voltage_charge': voltage_charge,
    'Time': time_data
})

# Print the DataFrame preview
display(df.head())
```
